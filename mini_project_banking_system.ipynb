```python
login_status=0
bal=100
def login(uname,pwd):
  if uname=='fasil' and pwd=='123':
    print('Login is successful')
    return 1
  else:
    print('Login is failed, please try again.....')
    return 0

def balance():
  global bal
  print("BALANCE: ",bal)

def deposite(amt):
  global bal
  bal=bal+amt
  print(amt," is successfully deposited")

def withdrawel(amt):
  global bal
  if bal>amt:
    bal=bal-amt
    print(amt," is successfully withdrawed")
  else:
    print("Withdrawel is failed due to insufficient balance")
```

```python
print("Welcome to MyBank -Login")
uname=input('Enter username:')
upwd=input('Enter password:')
login_status=login(uname,upwd)
while(login_status==1):
  str='''
  Please choose your service from the below list:
  1.Amount Depositing
  2.Amount Withdrawal
  3.Check Balance
  4.logout
  '''
  print(str)
  n=eval(input('Please choose any option:'))
  if n==1:
    dep_amt=eval(input('Please enter the amount for Depositing:'))
    deposite(dep_amt)
  elif n==2:
    wid_amt=eval(input('Please enter the amount for Withdrawal:'))
    withdrawel(wid_amt)
  elif n==3:
    print("Your balance is :",bal)
  elif n==4:
    print("Successfully logged out")
    login_status=0
  else:
    print("Wrong Command!")
```
